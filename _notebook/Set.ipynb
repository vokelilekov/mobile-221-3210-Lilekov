# Множество (Set)

```
//HashSet - неупорядоченное множество
//SortedSet - упорядоченное множество
```

```
HashSet<int> numbers = new();
numbers.Add(3);
numbers.Add(2);
numbers.Add(3);
numbers.Add(3);
numbers.Add(2);
numbers
```

```
// Проверка наличия элемента
numbers.Contains(3)
```

```
HashSet<int> a = new() {1, 2, 3, 4};
HashSet<int> b = new() {3, 4, 5, 6};

// Объединение множества
a.UnionWith(b);
a
```

```
HashSet<int> a = new() { 1, 2, 3, 4 };
HashSet<int> b = new() { 3, 4, 5, 6 };

// Пересечение множеств
a.IntersectWith(b);
a
```

```
HashSet<int> a = new() { 1, 2, 3, 4 };
HashSet<int> b = new() { 3, 4, 5, 6 };

// Разность множеств
a.ExceptWith(b);
a
```

```
HashSet<int> a = new() { 1, 2, 3, 4 };
HashSet<int> b = new() { 3, 4, 5, 6 };

// Симметричная разность множеств (только уникальные элементы)
a.SymmetricExceptWith(b);
a
```

```
HashSet<int> a = new() { 1, 2, 3, 4 };
HashSet<int> b = new() { 3, 4 };

bool isSubset = b.IsSubsetOf(a); // проверяет, является ли текущее множество подмножеством другого множества
bool isSuperset = a.IsSupersetOf(b); // проверяет, является ли текущее множество надмножеством другого множества.
bool hasOverlap = a.Overlaps(b); // проверяет, есть ли общие элементы между текущим и другим множеством.
bool areEqual = a.SetEquals(b); // проверяет, содержат ли множества одинаковые элементы.
isSubset
// isSuperset
// hasOverlap
// areEqual
```
